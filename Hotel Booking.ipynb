## Hotel Booking Analysis


```python
#inporting basic libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
sb.set()
import warnings
warnings.filterwarnings("ignore")
```

```python
#reading the hotel booking data
data=pd.read_csv("hotel_bookings.csv")
data.head()
```

```python
#Analysing the size
data.shape
```

```python
#getting datatype of all columns
data.describe()
```

```python
#checking the null values if any
data.isnull().sum()
```

```python
#filling null/missing values
data.fillna({'country':'unknown','children':0,'agent':0},inplace=True)
```

```python
#droping company as their is very lss data
data.drop('company',axis=1,inplace=True)
```

```python
#free of null valu3
data.isnull().sum()
```

```python
#checking and deleting duplicates
data.duplicated()
data=data.drop_duplicates()
```

```python
#checking correlation
plt.figure(figsize=(18,10))
sb.heatmap(data.corr(),annot=True,cmap="Reds",linecolor="white")
```

### Checking Features Visually

```python
data.hotel.value_counts()
```

```python
data
```

```python
#checking the types of hotel preffered mostly
data.hotel.value_counts().plot.bar(color="pink")
```

```python
data.lead_time
```

```python
#histogram to show how much time guest do the booking
plt.figure(figsize=(15,5))
data.lead_time.plot.hist(bins=41,edgecolor="blue")
plt.title("how much time guest do booking")
plt.xlabel("no of time")
```

```python
#cancellation rate
data.is_canceled.value_counts().plot.bar()
plt.xticks([0,1],['booked','cancel'])
```

```python
# Also, let's see how many cancellations are there in each hotel
sb.countplot(data.hotel,hue=data.is_canceled)
```

```python
data.groupby('hotel')['is_canceled'].value_counts(normalize=True)*100
```

```python
#price of hotel room
#removing outlier
data.adr.quantile([.70,.80,.90,.95,.99])
data=data[data.adr<261]
```

```python
#plotting adr
sb.boxplot(data.adr)
data.shape
```

```python
#prices round the year
sb.barplot(x=data.adr, y=data.arrival_date_month,hue=data.hotel)
```

Stay in Week nights v/s Stay in Weekend nights

```python
data.info()
```

```python
data.stays_in_week_nights.value_counts()
```

```python
a=data[data.stays_in_week_nights<10]
b=data[data.stays_in_weekend_nights<10]
```

```python
a.stays_in_week_nights.plot.hist(color='yellow',bins=8,label="weekdays")
b.stays_in_weekend_nights.plot.hist(bins=8,alpha=.5,label="weekends")
plt.legend(loc=1)
plt.xlabel("no. of days")
plt.ylabel("no of booking")
plt.title("no of booking in week and weekends")
```

```python
#Average Daily Price with respect to the room type
z1=data.groupby('reserved_room_type')['adr'].agg({'average_price':'mean','no of booking':'size'})

z1
```

```python
#visualizing above 
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(10,6))
ax1.plot(z1.average_price,color='red')

ax2.plot(z1['no of booking'],color='green')
```

```python
#Market Segment
plt.figure(figsize=(8,6))
data.market_segment.value_counts(sort=False).plot.pie(autopct='%1.1f%%',startangle=90)
```

```python
#Let's see from which countries the guests are bookings
(data.country.value_counts())[:10].plot.bar(color="green")
plt.title("top 10 countries")
```

```python
#Check which type of customers are visiting in each hotel
sb.countplot(data['customer_type'],hue=data.hotel)
```
